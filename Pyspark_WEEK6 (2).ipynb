### Q1: Demonstrate the Roles of Driver, Cluster Manager, and Executor in a Spark Application


```python

data = [
    (1, "Laptop", 65000),
    (2, "Mobile", 30000),
    (3, "Tablet", 25000),
    (4, "Monitor", 18000)
]
df = spark.createDataFrame(data, ["product_id", "product_name", "price"])

filtered_df = df.filter(df.price > 25000)
filtered_df.show()
```


#### Code Explanation

- We first create a Spark DataFrame using the sample data.
- Then, we apply the `filter()` transformation to select products with a price greater than **25,000**.
- Since `filter()` is a transformation, Spark only records this operation and does not execute it immediately.
- When we call `show()`, Spark starts executing the job.
- The Driver prepares the execution plan (DAG) and sends the tasks to the Executors with the help of the Cluster Manager.
- The Executors process the data and return the result to the Driver.
- Finally, the Driver displays the filtered records.


### Q2: Demonstrate Spark's Lazy Evaluation


```python
data = [
    (1, "Laptop", 65000),
    (2, "Mobile", 30000),
    (3, "Tablet", 25000),
    (4, "Monitor", 18000)
]

df = spark.createDataFrame(data, ["product_id", "product_name", "price"])
result_df = (
    df.filter(df.price > 20000)
      .select("product_name", "price")
      .withColumn("discount_price", df.price * 0.9)
)
result_df.show()
```

#### Code Explanation

- `filter()`, `select()`, and `withColumn()` are **Transformations**.
- Spark stores these operations but does **not** execute them immediately.
- `show()` is an **Action** that triggers the execution of all transformations together.
- Spark builds an optimized **DAG** and executes it efficiently.


Spark's Lazy Evaluation improves performance by delaying execution, optimizing the execution plan, and processing the data only when an action is invoked.

### Q4: What is the difference between CSV and Parquet in terms of storage (row-based vs. columnar) and why does it matter for performance? 

- CSV stores data **row by row**.
- Parquet stores data **column by column**.
- When we need only a few columns, Spark reads only those columns from a Parquet file, making it faster and more efficient for large datasets.

```python
csv_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/practice_catalog/new_source/new_volumeforcei/Sales.csv")

csv_df.write.mode("overwrite").parquet("/Volumes/practice_catalog/new_source/new_volumeforcei/Sales_Parquet")
parquet_df = spark.read.parquet("/Volumes/practice_catalog/new_source/new_volumeforcei/Sales_Parquet")

print("CSV Data")
csv_df.limit(5).display()
print("Parquet Data")
parquet_df.limit(5).display()
```

Both DataFrames display the same records because they contain the same data.

#### Code Explanation

- We first read the CSV file into a DataFrame.
- Next, we save the same DataFrame in Parquet format.
- Finally, we read the Parquet file and display both DataFrames.
- Although the output is the same, the internal storage format is different.

#### Comparison

| CSV | Parquet |
|------|----------|
| Row-based storage | Columnar storage |
| Larger file size | Smaller file size |
| Reads all columns | Reads only required columns |
| Slower for large datasets | Faster for analytics |
| Schema is not stored | Schema is stored with the file |

#### Conclusion

From this practical, we can understand that both formats store the same data. However, Parquet is more suitable for Spark because it stores data column-wise, uses less storage, and improves query performance when working with large datasets.

### Q7: How does Spark use the Lineage Graph (DAG) to provide fault tolerance if a worker node fails? 

- We perform multiple transformations on the DataFrame.
- Spark records these transformations as a **Lineage Graph (DAG)**.
- If an Executor fails, Spark uses the DAG to recompute only the lost partition instead of processing the entire dataset again.

```python
from pyspark.sql.functions import col

result_df = (
    df.filter(col("Item_MRP") > 100)
      .select("Item_Identifier", "Item_Type", "Item_MRP")
      .withColumnRenamed("Item_MRP", "Product_Price")
)
result_df.limit(5).show()
result_df.explain()
```


- We create a chain of transformations using `filter()`, `select()`, and `withColumnRenamed()`.
- These operations are stored as a **Lineage Graph (DAG)** and are not executed immediately.
- When `show()` is called, Spark executes the DAG.
- `explain()` displays the execution plan generated by Spark.

#### Conclusion

We can see that Spark keeps track of every transformation in the Lineage Graph. If an Executor fails while processing data, Spark uses this DAG to recompute only the missing partition instead of restarting the entire job. This provides fault tolerance and makes Spark reliable for processing large datasets.

### Q12: Write the Spark command to load a Parquet file from "path/to/input", filter out any rows where user_id is null, and save the result as a CSV at "path/to/output".
- user_id → Item_Identifier

```python
from pyspark.sql.functions import *

df = spark.read.parquet("/Volumes/practice_catalog/new_source/new_volumeforcei/Sales_Parquet")

filtered_df = df.filter(col("Item_Identifier").isNotNull())

filtered_df.write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv("/Volumes/practice_catalog/new_source/new_volumeforcei/Filtered_Sales_CSV")
```


- We read the Parquet file using `spark.read.parquet()`.
- We use `isNotNull()` to remove rows with null values in **Item_Identifier**.
- We save the filtered DataFrame as a CSV file using `write.csv()`.
- `header=true` writes the column names, and `overwrite` replaces the existing output if it already exists.

#### Conclusion

We created a simple data pipeline that reads a Parquet file, filters the data, and stores the processed result in CSV format.
